```python
from FICKETT_Class import *
import pandas as pd
from Bio import SeqIO
```

```python
cd ../CLASSES/FICKETT_Features
```

```python

input_file = "./input/sample100seq.fa"
fickett_out  = "./output/Fickett_Features.csv"

fasta_sequences = SeqIO.parse(open(input_file),'fasta')

with open(fickett_out, "w", encoding='utf-8') as fickett_f:
    header = ",".join(["id","A-base_position_score", "C-base_position_score", "G-base_position_score", "T-base_position_score"])
    fickett_f.write(header)
    fickett_f.write("\n")
    print("id,","A-base_position_score,", "C-base_position_score,", "G-base_position_score,", "T-base_position_score" )
    for fasta in fasta_sequences:
        seqid, sequence = fasta.id, str(fasta.seq)
        Fic = Fickett(seqid, sequence)
        Feature = Fic.gen_fic_fea()
        #Feature = Feature.replace("\t", ",")
        fic_fea = ",".join([str(seqid), str(Feature[0]), str(Feature[1]), str(Feature[2]), str(Feature[3])]) 
        fickett_f.write(fic_fea)  
        fickett_f.write("\n")
        print(Fic.id, ",", Feature[0], ",", Feature[1], ",", Feature[2], ",", Feature[3])
```
